```python
import pandas as pd
df=pd.read_csv(r"C:\Users\Percy\Downloads\adult.data.csv")


print(df.head())
print(df.info())



def calculate_demographic_data(print_data=True):
    # Read data from file
    df = pd.read_csv('/Users/mac/Downloads/adult.data.csv')

    # How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
    race_count = df['race'].value_counts()

    # What is the average age of men?
    average_age_men = df.groupby('sex')['age'].mean(['Male'])

    # What is the percentage of people who have a Bachelor's degree?
    number_of_bachelors = df['education'].value_counts()['Bachelors']
    total_count = len(df['education'])
    percentage_bachelors = (number_of_bachelors/total_count)*100

    # What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
    number_of_Masters= df['education'].value_counts()['Masters']
    number_of_doctorate= df['education'].value_counts()['Doctorate']
    higher_education = (number_of_bachelors + number_of_Masters + number_of_doctorate)
    counts = df.groupby(['education', 'salary']).size()
    bachelors_more_50K = counts.loc['Bachelors', '>50K']
    Masters_more_50K = counts.loc['Masters', '>50K']
    doctorate_more_50K = counts.loc['Doctorate', '>50K']
    Total_advanced_education_50K= bachelors_more_50K + Masters_more_50K + doctorate_more_50K
    Percentage_of_advanced_more_50K =(Total_advanced_education_50K/higher_education)*100

    # What percentage of people without advanced education make more than 50K?
    lower_education = (total_count - higher_education)
    Total_of_people_withsalary_more_50K = df['salary'].value_counts()['>50K']
    People_without_advanced_education_more_50K = (Total_of_people_withsalary_more_50K - Total_advanced_education_50K)
    lower_education_rich = (People_without_advanced_education_more_50K / lower_education) * 100

    # with and without `Bachelors`, `Masters`, or `Doctorate`
    higher_education = (number_of_bachelors + number_of_Masters + number_of_doctorate)
    lower_education = (total_count - higher_education)

    # percentage with salary >50K
    higher_education_rich = (Total_advanced_education_50K/higher_education)*100
    lower_education_rich = (People_without_advanced_education_more_50K / lower_education) * 100

    # What is the minimum number of hours a person works per week (hours-per-week feature)?
    min_work_hours = df['hours-per-week'].min()

    # What percentage of the people who work the minimum number of hours per week have a salary of >50K?
    num_min_workers = df['hours-per-week'].value_counts().get(1,0)

    rich_min_workers = sum(1 for i, j in zip(df['hours-per-week'], df['salary']) if i == 1 and j == '>50K')
    rich_percentage =(rich_min_workers/num_min_workers)*100


    # What country has the highest percentage of people that earn >50K?

    df_filtered = df[df['salary'] == '>50K']
    countries_over50K = df_filtered.groupby('native-country')['native-country'].count()
    grouped_total = df.groupby('native-country')['native-country'].count()
    percentages = (countries_over50K / grouped_total) * 100
    sorted_percentages = percentages.sort_values(ascending=False)
    highest_earning_country = sorted_percentages.index[0]

    highest_earning_country_percentage = sorted_percentages.iloc[0]

    # Identify the most popular occupation for those who earn >50K in India.
    df_india= df[df['native-country']=='India']
    df_india_over_50K= df_india[df_india['salary']=='>50K']
    grouped_occupation= df_india_over_50K.groupby('occupation')['occupation'].count()
    sorted_occupation = grouped_occupation.sort_values(ascending=False)
    top_IN_occupation = sorted_occupation.index[0]

    # DO NOT MODIFY BELOW THIS LINE

    if print_data:
        print("Number of each race:\n", race_count)
        print("Average age of men:", average_age_men)
        print(f"Percentage with Bachelors degrees: {percentage_bachelors}%")
        print(f"Percentage with higher education that earn >50K: {higher_education_rich}%")
        print(f"Percentage without higher education that earn >50K: {lower_education_rich}%")
        print(f"Min work time: {min_work_hours} hours/week")
        print(f"Percentage of rich among those who work fewest hours: {rich_percentage}%")
        print("Country with highest percentage of rich:", highest_earning_country)
        print(f"Highest percentage of rich people in country: {highest_earning_country_percentage}%")
        print("Top occupations in India:", top_IN_occupation)

    return {
        'race_count': race_count,
        'average_age_men': average_age_men,
        'percentage_bachelors': percentage_bachelors,
        'higher_education_rich': higher_education_rich,
        'lower_education_rich': lower_education_rich,
        'min_work_hours': min_work_hours,
        'rich_percentage': rich_percentage,
        'highest_earning_country': highest_earning_country,
        'highest_earning_country_percentage':
            highest_earning_country_percentage,
        'top_IN_occupation': top_IN_occupation
    }
```

```python
# Call the function and pass print_data as True to print the results
results = calculate_demographic_data(print_data=True)

# You can access the results using the dictionary keys
race_count = results['race_count']
average_age_men = results['average_age_men']
percentage_bachelors = results['percentage_bachelors']
higher_education_rich = results['higher_education_rich']
lower_education_rich = results['lower_education_rich']
min_work_hours = results['min_work_hours']
rich_percentage = results['rich_percentage']
highest_earning_country = results['highest_earning_country']
highest_earning_country_percentage = results['highest_earning_country_percentage']
top_IN_occupation = results['top_IN_occupation']
```
